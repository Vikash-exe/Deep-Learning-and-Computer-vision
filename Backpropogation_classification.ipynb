```python
def initialize_parameters(layer_dims):
    np.random.seed(12)
    parameters = {}
    for l in range(1, len(layer_dims)):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.1 #matrix of current layer and previous layer for weights
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1)) #matrix of current bias layer and 1 column

    return parameters
```

```python
parameters = initialize_parameters([2, 2, 1])
parameters
```

```python
#sigmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def linear_forward(a_prev, W, b):
    z = np.dot(W, a_prev) + b
    A = sigmoid(z)
    return A
```

```python
def L_linear_forward(x, parameters):
    A = x
    L = len(parameters) // 2  #number of layers in the neural network is half the number of parameters because we have W and b for each layer
    for l in range(1, L+1):
        A_prev = A
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        A = linear_forward(A_prev, W, b)

    return A, A_prev
```

```python
def update_parameters(parameters, y , y_hat, A1, x):
    parameters['W2'][0][0] = parameters['W2'][0][0] + ( 0.001 * (y - y_hat) * A1[0][0])
    parameters['W2'][0][1] = parameters['W2'][0][1] + ( 0.001 * (y - y_hat) * A1[1][0])
    parameters['b2'][0][0] = parameters['b2'][0][0] + ( 0.001 * (y - y_hat) * 1)

    parameters['W1'][0][0] = parameters['W1'][0][0] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0]*(1-A1[0][0]) * x[0][0])
    parameters['W1'][1][0] = parameters['W1'][1][0] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0]*(1-A1[0][0]) * x[1][0])
    parameters['b1'][0][0] = parameters['b1'][0][0] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0]*(1-A1[0][0]) * 1)

    parameters['W1'][1][0] = parameters['W1'][1][0] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[1][0]*(1-A1[1][0]) * x[0][0])
    parameters['W1'][1][1] = parameters['W1'][1][1] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[1][0]*(1-A1[1][0]) * x[1][0])
    parameters['b1'][1][0] = parameters['b1'][1][0] + ( 0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[1][0]*(1-A1[1][0]) * 1)
    return parameters
```

```python
x = df[['Feature1', 'Feature2']].values[0].reshape(2,1)
y = df[['Target']].values[0][0]
parameters = initialize_parameters([2, 2, 1])
y_hat, A1 = L_linear_forward(x, parameters)
y_hat = y_hat[0][0]

update_parameters(parameters, y, y_hat, A1, x)

print('loss for first data point: ', -y*np.log(y_hat) - (1-y)*np.log(1-y_hat))
parameters
```

```python
#epoch implementation for decent learning

parameters = initialize_parameters([2, 2, 1])
for epoch in range(1000):
    loss = []
    for i in range(len(df)):
        x = df[['Feature1', 'Feature2']].values[i].reshape(2,1)
        y = df[['Target']].values[i][0]
        y_hat, A1 = L_linear_forward(x, parameters)
        y_hat = y_hat[0][0]

        parameters = update_parameters(parameters, y, y_hat, A1, x)
        loss.append(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))
    print('epoch: ', epoch, 'loss: ', np.mean(loss))
```
