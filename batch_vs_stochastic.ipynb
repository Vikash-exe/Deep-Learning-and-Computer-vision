```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import Adam
```

```python
df = pd.read_csv(r"C:\Users\VIKASH KUMAR\Downloads\DATA_DUMP\archive (5)\Admission_Predict.csv")
```

```python
df.head()
```

```python
df.drop(['Serial No.', 'Research'], axis=1, inplace=True)
```

```python
df.head()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Sequential([
    Input(shape=(6,)),
    Dense(10, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
x_train.shape
```

```python
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['mae'])
import time
start_time = time.time()
history = model.fit(x_train, y_train, epochs=100, batch_size=320, validation_data=(x_test, y_test)) # Batch Gradient Descent and 320 is the total number of samples in the training set
end_time = time.time()
print(f"Training time: {end_time - start_time:.2f} seconds")
```

```python
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['mae'])
import time
start_time = time.time()
history = model.fit(x_train, y_train, epochs=100, batch_size=1, validation_data=(x_test, y_test)) # Stochastic Gradient Descent
end_time = time.time()
print(f"Training time: {end_time - start_time:.2f} seconds")
```
