```python
def initialize_parameters(layer_dims):
    np.random.seed(12)
    parameters = {}
    for l in range(1, len(layer_dims)):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.1 #matrix of current layer and previous layer for weights
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1)) #matrix of current bias layer and 1 column

    return parameters
```

```python
parameters = initialize_parameters([2, 2, 1])
parameters
```

```python
def linear_forward(A_prev, W, b):
    z = np.dot(W, A_prev) + b
    return z
```

```python
def L_layer_forward(x,parameters):
    a = x
    l = len(parameters) // 2

    for i in range(1, l+1):
        a_prev = a
        wl = parameters['W' + str(i)]
        bl = parameters['b' + str(i)]
        # print("A"+str(i-1)+ ": ", a_prev)
        # print("W"+str(i)+ ": ", wl)
        # print("b"+str(i)+ ": ", bl)
        # print("--"*15)

        a = linear_forward(a_prev, wl, bl)
        # print("a"+str(i)+ ": ", a)
        # print("=="*15)

    return a, a_prev
```

```python
x = df[['Feature1', 'Feature2']].values[0].reshape(2,1) # Convert to column vector
y = df['Target'].values[0].reshape(1,1) # Convert to column vector
```

```python
def update_paramters(parameters, y , y_hat, a1, x):
    error = (y - y_hat).item()
    parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (error) * a1[0][0])
    parameters['W2'][0][1] = parameters['W2'][0][1] + (0.001 * 2 * (error) * a1[1][0])
    parameters['b2'][0][0] = parameters['b2'][0][0] + (0.001 * 2 * (error) * 1)

    parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (error) * parameters['W2'][0][0] * x[0][0])
    parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (error) * parameters['W2'][0][0] * x[1][0])
    parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (error) * parameters['W2'][0][0] * 1)

    parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (error) * parameters['W2'][0][1] * x[0][0])
    parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (error) * parameters['W2'][0][1] * x[1][0])
    parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (error) * parameters['W2'][0][1] * 1)
    return parameters
    
```

```python
#epoch implementation
parameters = initialize_parameters([2, 2, 1])
epochs = 10

for i in range(epochs):
    loss = [0]

    for j in range(df.shape[0]):
        x = df[['Feature1', 'Feature2']].values[j].reshape(2,1) # Convert to column vector
        y = df['Target'].values[j].reshape(1,1) # Convert to column vector

        y_hat, A1 = L_layer_forward(x, parameters)
        y_hat = y_hat[0][0]
        loss.append(((y - y_hat) ** 2).item())
        parameters = update_paramters(parameters, y , y_hat, A1, x)
    print('Epoch: ', i+1, ' Loss: ', np.array(loss).mean())
print(parameters)
```
